Vamos definir duas funções para ser utilizadas para gerar alguns gráficos:

```python
def show_plot1(var, labels):
    (df.groupby('Diabetes_binary')[var]
     .value_counts(1)
     .unstack()
     .iloc[:,::-1]
     .plot(kind='barh', stacked=True,figsize=(12, 4), color=sns.color_palette('bright'), alpha=1)
    )

    plt.legend(labels,bbox_to_anchor=(1, 1, 0, 0), shadow=False, frameon=False)
    plt.yticks(ticks=[0,1], labels=['Non-Diabetic', 'Diabetic'])
    plt.title(var, fontsize=16)
    plt.show()
```

```python
def show_plot2(x, y, title):

  df1 = data.groupby(x)[y].value_counts(normalize=True)
  df1 = df1.mul(100)
  df1 = df1.rename('percent').reset_index()

  g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df1, palette=sns.color_palette('bright'))
  g.ax.set_ylim(0,100)

  for p in g.ax.patches:
      txt = str(p.get_height().round(2)) + '%'
      txt_x = p.get_x()
      txt_y = p.get_height()
      g.ax.text(txt_x, txt_y, txt)

  plt.title(title, size = 12);
```

Vamos definir uma função que mostra os principais parâmetros importantes na avaliação de um algoritmo, para simplificar o código:

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error
import math

def gen_confusion_matrix(y_test, y_pred):

  labels = ["True Neg", "False Pos", "False Neg", "True Pos"]
  categories = ["No diabetes", "Diabetes"]

  cm = confusion_matrix(y_test, y_pred)

  sns.heatmap(cm/np.sum(cm), annot=True, annot_kws={"fontsize":12}, fmt='.2%', cmap='BuPu', cbar=True,
              xticklabels=categories, yticklabels = categories)
  sns.set(font_scale=1)
  plt.xlabel('Predicted label', fontsize=12)
  plt.ylabel('True label', fontsize=12)
  plt.title('Matriz de Confusão', fontsize=14)
  plt.show()

def show_results(y_test, y_pred):

  print("A acurácia foi:", accuracy_score(y_test, y_pred))
  print('Mean Squared Error : ', mean_squared_error(y_test, y_pred))
  print('Root Mean Squared Error : ', math.sqrt(mean_squared_error(y_test, y_pred)))

  print('-----------')
  print(classification_report(y_test,y_pred))

  gen_confusion_matrix(y_test, y_pred)
```

### ***5.6 Random Forest***

As florestas aleatórias (Random Forest), como o nome sugere, utilizam uma combinação de árvores de decisão para realizar tarefas de machine learning. Sua implementação depende de hiperparâmetros, logo também podemos utilizar a validação cruzada para o ajuste do modelo:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

mauc = []
macc = []
vn = []

for n in range(1, 50):
    vauc = []
    vacc = []
    for train_index, test_index in cv.split(x_train, y_train):

        x_t, x_validation = x_train[train_index],x_train[test_index]
        y_t, y_validation = y_train[train_index], y_train[test_index]

        model = RandomForestClassifier(n_estimators=n)
        model.fit(x_t,y_t)
        y_pred = model.predict(x_validation)
        score = accuracy_score(y_pred, y_validation)

        vauc.append(roc_auc_score(y_validation, y_pred))
        vacc.append(accuracy_score(y_pred, y_validation))
    macc.append(np.mean(vacc))
    mauc.append(np.mean(vauc))
    vn.append(n)
best_n = np.argmax(mauc)+1
print('Melhor n:', best_n, ' AUC:',mauc[best_n-1])
plt.figure(figsize=(10,5))
plt.plot(vn, mauc, '-ro', label= 'AUC')
plt.plot(vn, macc, '-bo', label = 'Accuracy')
plt.xlabel('n', fontsize = 15)
plt.ylabel('Score', fontsize = 15)
plt.legend()
plt.show()
```

```python
model = RandomForestClassifier(n_estimators=best_n)
model.fit(x_train, y_train)

y_pred_rf = model.predict(x_test)

show_results(y_test, y_pred_rf)
```

É notável que as perguntas relacionadas com saúde e IMC possuem maior importância na classificação. Logo em seguida, vemos que a renda e idade também foram características importantes para o classificador. Vamos tentar utilizar as 4 primeiras e avaliar como o desempenho do modelo varia:

```python
ns = 4 #numero de atributos selecionados
selected_attributes = []

for i in range(0,ns):
    selected_attributes.append(attributes_rank[::-1][i])
print('Selecionados:', selected_attributes)

x_selected = df[selected_attributes]
x_selected = x_selected.to_numpy()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)

mauc = []
macc = []
vn = []
for n in range(1, 50):
    vauc = []
    vacc = []
    for train_index, test_index in cv.split(x_train, y_train):

        x_t, x_validation = x_train[train_index], x_train[test_index]
        y_t, y_validation = y_train[train_index], y_train[test_index]

        model = RandomForestClassifier(n_estimators=n)
        model.fit(x_t, y_t)
        y_pred = model.predict(x_validation)
        score = accuracy_score(y_pred, y_validation)

        vauc.append(roc_auc_score(y_validation, y_pred))
        vacc.append(accuracy_score(y_pred, y_validation))
    macc.append(np.mean(vacc))
    mauc.append(np.mean(vauc))
    vn.append(n)
best_n = np.argmax(mauc)+1
print('Melhor n:', best_n, ' AUC:',mauc[best_n-1])
plt.figure(figsize=(10,5))
plt.plot(vn, mauc, '-ro', label= 'AUC')
plt.plot(vn, macc, '-bo', label = 'Accuracy')
plt.xlabel('n', fontsize = 15)
plt.ylabel('Score', fontsize = 15)
plt.legend()
plt.show()
```

Iniciamos com um dataset baseado em uma pesquisa feita por telefone para moradores dos Estados Unidos que visava encontrar relações entre inúmeros fatores (não invasivos, que não precisavam de exames clínicos, como seria o caso de solicitar níveis de insulina) e a prevalência de casos de Diabetes. Nela haviam perguntas como idade, sexo, salário, a frequência com que praticava atividade física, dentre outras.

Com base nos resultados das matrizes de confusão, o algoritmo mais adequado nesse dataset é o Random Forest, uma vez que este gerou um modelo final mais forte e preciso do que os outros comparativamente. Random Forest é um dos algoritmos de Ensemble que combina diversas Decision Trees ao longo do treinamento com os dados, combinado com outras técnicas de melhoria de hiperparâmetros (como o melhor número de árvores a ser considerada) conseguimos chegar a uma acurácia de 90%, ou seja, 9 a cada 10 diagnósticos estariam corretos se feitos inteiramente por machine learning. Dos diagnósticos incorretos, praticamente todos seriam falsos negativos. Não é recomendado o diagnóstico por algoritmos de aprendizado, mas havendo sempre um acompanhamento médico esta taxa de acerto pode ser útil para auxiliar o profissional no tratamento de seus pacientes. Não apenas isso, mas os algoritmos servem também para retornar quais atributos podem ser mais importantes a serem levados em consideração, que dá ao profissional um norte de por onde começar a investigar possíveis causas.

\\

De maneira geral, o estudo foi importante para ressaltar a sua habilidade em ser utilizado como referência inicial para o auxílio do profissional da saúde, neste contexto, o diagnóstico de diabetes, mas mostra potencial para o uso em outras áreas.
