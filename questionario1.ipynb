### ***Questão 02***

*Leia os dados com erro novamente e insira a média de cada atributo onde for encontrada `NaN` e `'?'`. Quais os valores das medianas obtidas para os atributos?*

```python
data = pd.read_csv('data/iris-with-errors.csv', header=(0))
data.head(25)
```

```python
data = data.replace('?', np.nan)
data.head(25)
```

### Questão 03

*Monte um boxplot para cada variável dos dados da Iris. Quais atributos apresentam outliers?*

```python
import seaborn as sns

sns.set_style('darkgrid')

def show_boxplot(data, attribute, title):

  data = pd.read_csv('data/iris.csv', header=(0))
  plt.figure(figsize=(6, 6))

  sns.boxplot(x="species", y=attribute, data=data)

  plt.xlabel('Espécie', fontsize=12)
  plt.ylabel(title, fontsize=12)
  plt.xticks(fontsize=10)
  plt.yticks(fontsize=10)

  plt.show()
```

```python
show_boxplot(data, 'sepal_length', 'Comprimento da sépala')
```

```python
show_boxplot(data, 'sepal_width', 'Largura da sépala')
```

```python
show_boxplot(data, 'petal_length', 'Comprimento da pétala')
```

```python
show_boxplot(data, 'petal_width', 'Largura da pétala')
```

Vemos através dos diagramas que todas as variáveis apresentam pelo menos um outlier em pelo menos uma das espécies.

### ***Questão 04***

*Considere a base de dados do Titanic, disponível na plataforma Kaggle. Calcule a correlação entre as cariáveis. Quais são as variáveis mais correlacionadas?*

Importando o dataset:

```python
import requests
from io import StringIO
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

response = requests.get(url)
data = response.text
```

```python
df = pd.read_csv(StringIO(data))

df.head(5)
```

Retirando atributos como nome e id:

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

Conversão de variáveis nominais em inteiros:

```python
df = df.replace('S', 0)
df = df.replace('C', 1)
df = df.replace('Q', 2)
```

```python
df = df.replace('female', 0)
df = df.replace('male', 1)
```

Calculando a matriz de correlação:

```python
corr = df.corr()
corr
```

Visualizando-a:

```python
plt.figure(figsize=(7, 7))
plt.imshow(corr, cmap='PuRd', interpolation='none', aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

Já podemos visualizar na matriz as variáveis com maior correlação. Para verificar e obter os valores mais positivos:

```python
var_list = []

for i in corr.columns:
    for j in corr.columns:
        if (i > j):
          var_list.append([i, j, corr[i][j]])

var_ord = sorted(var_list, key=lambda x: x[-1], reverse = True)

print('Variáveis com maior correlação:')
print(f'{var_ord[0][0]} e {var_ord[0][1]}: {var_ord[0][-1]}')
print(f'{var_ord[1][0]} e {var_ord[1][1]}: {var_ord[1][-1]}')
```
